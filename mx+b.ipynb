```python
# Single line using y = mx + b
```

```python
x = [-5,5]
m = 2
b = 1

y = [0,0]


for i in range(0,len(x)):
    y[i] = m*x[i] + b
    
print(x,y)
```

```python
# Two lines using y = mx + b
```

```python
x = [-5,5]
m = [0.7,-1.25]
b = [-2,0.75]


for i in range(0, len(m)):
    y = m[i]*np.array(x) + b[i]

print(x,y)
```

```python
# Graphin rational functions
```

```python
x = range(-3,4)
y = np.zeros(len(x))


for i in range(0, len(x)):
    y[i] = 2 - x[i]**2
    
print(x,y)
```

```python
plt.plot(x,y,'s-')
```

```python
x = np.linspace(-3,3, 10)
y = 2 + np.sqrt(abs(x))
```

```python
plt.plot(y)
```

```python
x = np.linspace(-4,4,4)
e = range(-1,4)
for i in e:
    y = x**i
    plt.plot(x,y)
```
