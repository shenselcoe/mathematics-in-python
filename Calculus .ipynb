```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
x = sym.symbols('x')
y = sym.symbols('y')
```

```python
# find the Limit
```

```python
fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx,x,lim_pnt)
print(lim_pnt,fx,lim)
```

```python
fxx = sym.lambdify(x,fx)
xx = np.linspace(-5,5,200)

plt.plot(xx,fxx(xx))
```

```python
fx = (x**2)/(x-2)
fxx = sym.lambdify(x,fx)
xx = np.linspace(1,3,102)

lim_pnt = 2
lim = sym.limit(fx,x,lim_pnt, dir = '-')

plt.plot(xx,fxx(xx))
lim
```

```python
# Piecewise functions
```

```python
piece1 = 0 
piece2 = -2*x
piece3 = x**3 /10
```

```python
fx = sym.Piecewise((piece1,x <0), (piece2, (x >=0) & (x<10)),(piece3, x >= 10) )
fx

fxx = sym.lambdify(x,fx)
xx = np.linspace(-3,15,1234)
plt.plot(xx,fxx(xx))
```

```python
import sympy.plotting.plot as symplot

fx = 3 - x**3
p = symplot(fx,(x,-5,5),show = False)
p.extend(symplot(sym.diff(fx),(x,-5,5),show = False))
p.show()
```
